```python
# prepare classification dataset
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)

# plot the data
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
import numpy as np

# define a perceptron and find it's weights using Activation function = step function and Loss function = Hinge loss
def perceptron(X, y):
    epochs = 1000
    lr = 0.1

    # Add 1 to start of X for simplification of calculations
    X = np.insert(X, 0, 1, axis=1)

    # initialize the weights
    w = np.ones(X.shape[1])

    # Apply stochastic gradient descent
    for i in range(epochs):
        # Iterate over each training data and apply gradient descent
        for j in range(X.shape[0]):
            zj = np.dot(w, X[j])

            if y[j]*zj < 0:
                w = w - lr * np.dot(y[j], X[j])
    
    return w

# def perceptron(X,y):
    
#     w1=w2=b=1
#     lr = 0.1
    
#     for j in range(1000):
    
#         for i in range(X.shape[0]):

#             # check condition
#             z = w1*X[i][0] + w2*X[i][1] + b

#             if z*y[i] < 0:
#                 w1 = w1 - lr*y[i]*X[i][0]
#                 w2 = w2 - lr*y[i]*X[i][1]
#                 b = b - lr*y[i]
            
#     return [b,w1,w2]
            
```

```python
w = perceptron(X, y)
m = -1 * w[1] / w[2]
b = -1 * w[0] / w[2]
print("m =", m)
print("b =", b)
```

```python
# plot the data
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)

x_p = np.linspace(-3, 3, 100)
y_p = m * x_p + b
plt.plot(x_p, y_p, color='red')
plt.ylim(-3, 3)
```
