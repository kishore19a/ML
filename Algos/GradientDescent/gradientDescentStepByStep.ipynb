```python
# Gradient descent for linear regression
# prepare a dataset with 4 training data - 1 independent and 1 target variable

from sklearn.datasets import make_regression
import numpy as np

# generate regression dataset
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)

# plot regression dataset
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
# apply sklear linear regression (Internally uses OLS)
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# fit the model
model.fit(X, y)

# get the coeffiecients and intercept
m = model.coef_[0]
c = model.intercept_

# print the coefficients and intercept
print("Coefficients: ", m)
print("Intercept: ", c)

# plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', label='Linear Regression')
plt.legend()
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0

y_pred = ((78.35 * X) + 0).reshape(4)

# plot the original data, the regression line and the initial prediction
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='b = 0')
plt.legend()
```

```python
# Epoch 1

m = 78.35
b = 0
learningRate = 0.1

# calculate the step size
stepSize = learningRate * (-2 * np.sum(y - (m * X.ravel() + b)))
print("Step size: ", stepSize)

# update the intercept
b = b - stepSize
print("Updated intercept: ", b)

# calculate the new predictions
y_pred1 = ((78.35 * X) + b).reshape(4)

# plot the original data, the regression line and all the predictions
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='b = 0')
plt.plot(X, y_pred1, color='orange', label='b = {0:.3f}'.format(b))
plt.legend()
```

```python
# Epoch 2

# calculate the step size
stepSize = learningRate * (-2 * np.sum(y - (m * X.ravel() + b)))
print("Step size: ", stepSize)

# update the intercept
b = b - stepSize
print("Updated intercept: ", b)

# calculate the new predictions
y_pred2 = ((78.35 * X) + b).reshape(4)

# plot the original data, the regression line and all the predictions
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='b = 0')
plt.plot(X, y_pred1, color='orange', label='b = {0:.3f}'.format(b))
plt.plot(X, y_pred2, color='purple', label='b = {0:.3f}'.format(b))
plt.legend()
```

```python
# Epoch 3

# calculate the step size
stepSize = learningRate * (-2 * np.sum(y - (m * X.ravel() + b)))
print("Step size: ", stepSize)

# update the intercept
b = b - stepSize
print("Updated intercept: ", b)

# calculate the new predictions
y_pred3 = ((78.35 * X) + b).reshape(4)

# plot the original data, the regression line and all the predictions
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='b = 0')
plt.plot(X, y_pred1, color='orange', label='b = {0:.3f}'.format(b))
plt.plot(X, y_pred2, color='purple', label='b = {0:.3f}'.format(b))
plt.plot(X, y_pred3, color='brown', label='b = {0:.3f}'.format(b))
plt.legend()
```

We can see that with each iteration, the predicted line is moving towards the best fit line (as predicted by OLS)

```python
# Lets write a loop for each epoch
# initialize the variables
b = 0
m = 78.35
learningRate = 0.1
epochs = 10

# plot the original data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', label='OLS')

# loop through the number of epochs
for i in range(epochs):
    # calculate the step size
    stepSize = learningRate * (-2 * np.sum(y - (m * X.ravel() + b)))
    # update the intercept
    b = b - stepSize
    # calculate the new predictions
    y_pred = (m * X.ravel()) + b

    plt.plot(X, y_pred)
```
