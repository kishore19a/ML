```python
# import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load placement.csv file
df = pd.read_csv('placement.csv')
print(df.head())
```

```python
# plot df
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package in LPA')
```

```python
# create x and y
x = df['cgpa']
y = df['package']
```

```python
# split data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
# train simple linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
m = model.coef_
c = model.intercept_
print('Slope:', m)
print('Intercept:', c)
```

```python
print("Test x = ", x_test.iloc[0].reshape(1, -1))
y_pred = model.predict(x_test.iloc[0].reshape(1, -1))
print("Predicted y = ", y_pred)
print("Test y = ", y_test.iloc[0])
```

```python
# plot regression line
plt.scatter(x, y)
plt.plot(x_train, model.predict(x_train), color='red')
# write equation of line along with slope and intercept on the line itself
plt.text(6.5, 0, 'y = {}x + {}'.format(m, c), fontsize=15)
plt.xlabel('CGPA')
plt.ylabel('Package in LPA')
plt.show()
```
